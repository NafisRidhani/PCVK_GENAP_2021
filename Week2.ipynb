```
# Get Image from GD and read it using Matplotlib
import cv2 as cv
from google.colab.patches import cv2_imshow
from skimage import io
import matplotlib.pyplot as plt
import numpy as np

img = cv.imread('/content/drive/MyDrive/PCVK/kucing.jpg')
plt.imshow(img)
```

```
# Convert to RGB Color
img2 = img
img3 = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img3)
```

```
# RGB to Grayscale
img_gray = cv.imread('/content/drive/MyDrive/PCVK/kucing.jpg', cv.IMREAD_GRAYSCALE)
plt.imshow(img_gray)
```

```
# Gray colormap
plt.imshow(img_gray, cmap='gray')
```

```
# Magma Colormap
plt.imshow(img_gray, cmap='magma')
```

```
# Resize to 512 x 1024
img4 = cv.resize(cv.cvtColor(img, cv.COLOR_BGR2RGB), (512, 1024))
plt.imshow(img4)
```

```
# Resize and rotate
img5 = cv.flip(cv.cvtColor(img, cv.COLOR_BGR2RGB), 0)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.imshow(img5)
```

```
# Geometry 2D
background = np.zeros(shape=(512, 512, 3), dtype = np.int16)
plt.imshow(background)
```

```
# Green Rectangle
cv.rectangle(background, pt1=(384, 0), pt2=(510, 150), color = (0, 255, 0), thickness = 10)
plt.imshow(background)
```

```
# Blue Rectangle
cv.rectangle(background, pt1=(200, 200), pt2=(300, 300), color = (0, 0, 255), thickness = 15)
plt.imshow(background)
```

```
# Red Circle
cv.circle(background, center=(100,100), radius=50, color = (255, 0, 0), thickness = 10)

cv.circle(background, center=(400, 400), radius=50, color=(255,0,0), thickness=-1)

plt.imshow(background)
```

```
# Line
cv.line(background, pt1=(0, 0), pt2=(512, 512), color = (255, 255, 255), thickness = 10)
plt.imshow(background)
```

```
# Text
font = cv.FONT_HERSHEY_SIMPLEX
cv.putText(background, text='Hello', org=(10,500), fontFace=font, fontScale=4, color=(255, 255, 0), thickness=2, lineType=cv.LINE_AA)
plt.imshow(background)
```

```
# Canvas 2
bg = np.zeros(shape=(512, 512, 3), dtype=np.int32)
vertices = np.array([[100, 300], [200, 200], [400, 300], [200, 400]], dtype=np.int32)
pts = vertices.reshape((-1, 1, 2))
cv.polylines(bg,[pts], isClosed=True, color=(255, 255, 0), thickness=5)
plt.imshow(bg)
```
